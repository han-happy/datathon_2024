```python
# imports

import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load in data

tract_zip_dat = pd.read_csv('trackzip.csv') # data from US bureau to be used to convert censustract to zipcode

grocery_dat = pd.read_csv('grocery.csv') # data from US FDA 
```

```python
washed_grocery = grocery_dat.dropna(axis=1)
washed_grocery
```

```python
# Create a mapping dictionary from census tract to zip code using US_geo_dat
census_tract_to_zip = dict(zip(tract_zip_dat['TRACT'], tract_zip_dat['ZIP']))

# Map zip codes to grocery_dat based on matching census tract and census key
washed_grocery.loc[:, 'ZipCode'] = washed_grocery['CensusTract'].map(census_tract_to_zip)

washed_grocery.head()
```

```python
selected_columns = ['CensusTract', 'State', 'County', 'LILATracts_1And10', 'LILATracts_halfAnd10', 'LILATracts_1And20', 'LILATracts_Vehicle', 'ZipCode']

new_grocery = washed_grocery[selected_columns]
```
